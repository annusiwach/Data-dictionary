```python
import pandas as pd
import numpy as np
```

# Load the dataset

```python
df = pd.read_excel(r"C:\Users\annus\Downloads\ApexPlanet_DataAnalytics_Dataset.xlsx")
df.head()
```

```python
print(f"Dataset Shape: {df.shape}")
df.info()
```

```python
print(df.columns.tolist())
df.columns = df.columns.str.strip().str.lower()
df.columns = df.columns.str.replace(' ', '_')
print("Cleaned columns:", df.columns.tolist())

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
print(df['order_date'].dtype)
```

```python
median_age = df['age'].median()
print(f"Calculated Median Age: {median_age}")

df['age'] = df['age'].fillna(median_age)
df['age'] = df['age'].astype(int)
```

```python
df['city'] = df['city'].fillna("Unknown")
```

# Checking for missing values

```python
missing_counts = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) *100

missing_summary = pd.DataFrame({'Missing Values': missing_counts, 'Percentage (%)': missing_percentage})
print(missing_summary)
```

# Check for Duplicate Rows

```python
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows rows found: {duplicate_count}")

if duplicate_count > 0:
    display(df[df.duplicated(keep=False)].head())
```

# Verify Mathematical Integrity

```python
calculated_sales = df['quantity'] * df['unit_price']
math_mismatch = df[np.isclose(df['total_sales'], calculated_sales) == False]

print(f"Rows with mathematical discrepancies: {len(math_mismatch)}")
```
